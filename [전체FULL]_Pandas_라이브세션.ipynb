# 3-2) 시계열 처리

**▶ payments2 - "주문시간" datetime형태로 type 변환**

```python
payments2['주문시간'] = pd.to_datetime(payments2["주문시간"])
```

```python
payments2['주문시간']
```

**▶payments2에서 '주문시간' 칼럼을 이용해 연도만 추출하기 ("주문년도" 생성)**

```python
payments2['주문년도'] = payments2['주문시간'].dt.year
```

**▶payments2에서 '주문시간' 칼럼을 이용해 월만 추출하기 ("주문월" 생성)**

```python
payments2['주문월'] = payments2['주문시간'].dt.month
```

**▶payments2에서 '주문시간' 칼럼을 이용해 일만 추출하기 ("주문일" 생성)**

```python
payments2['주문일'] = payments2['주문시간'].dt.day
```

```python
payments2
```

**▶payments2에서 "주문연월" 컬럼을 생성해보자**

```python
# 첫번째 방법 (문자열 결합)
payments2['주문년도'].astype(str) + "-" + payments2['주문월'].astype(str)
```

```python
# 두번째 방법 (dt.strftime 이용)
payments2["주문연월"] = payments2["주문시간"].dt.strftime("%Y-%m")
```

# 3-3) 그룹화

**▶payments2에서 "결제방식"별 "결제금액" 평균을 구해보자 (pay_mean) 생성**

```python
pay_mean = payments2.groupby('결제방식')['결제금액'].mean()
```

**▶payments2에서 "결제방식"별 "결제금액" 평균, 합계 구해보자 (pay_groupby) 생성**

```python
pay_groupby = payments2.groupby('결제방식')['결제금액'].agg(["mean","sum"])
```

**▶인덱스를 컬럼으로 바꾸기**

```python
pay_mean = pay_mean.reset_index()
```

```python
pay_mean
```

```python
pay_groupby.reset_index(inplace = True)
```

```python
pay_groupby
```

# 3-4) 피봇팅

**▶payments2에서 주문연월별 주문상태 및 결제방식에 따른 평균 결제 금액, 고유 count을 알아보자. 총합계 출력 (result)**

```python
payments2
```

```python
result = payments2.pivot_table(values = "결제금액",
                      index = ["주문상태","결제방식"],
                      columns = ["주문연월"],
                      aggfunc = ["mean", "count"],
                      fill_value = 0,
                      margins = True)
```

```python
result.reset_index(inplace = True)
result
```
