```python
# 컬럼 정보 조회
# sql 해설에서 주문일자 값으로 order_purchase_timestamp 컬럼을 사용
orders.dtypes
```

```python
# 1단계:날짜 "%Y-%m" 형태로 변형
# 01. 해당 컬럼의 타입을 날짜열로 변환
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
```

```python
# 1단계:날짜 "%Y-%m" 형태로 변형
# 02. 날짜 포맷팅
import datetime as dt
orders["Year"] = orders['order_purchase_timestamp'].dt.year
orders['Month']= orders['order_purchase_timestamp'].dt.month
orders['order_purchase_timestamp'] = orders['order_purchase_timestamp'].dt.strftime('%Y-%m')
```

```python
# 데이터프레임 열 순서 정리
orders.insert(0, 'Year', orders.pop('Year'))
orders.insert(1, 'Month', orders.pop('Month'))
```

```python
# 데이터프레임 열 정리
orders = orders.loc[-1:, 'Year':'order_purchase_timestamp']
orders
```

```python
# 2단계: 월별 주문 건수 계산 및 누락 월 확인
# 01. 연월별 그룹 만들기 및 주문 건수 집계
orders_df = orders.groupby(['Year', 'Month'])['order_id'].count().reset_index()
orders_df

# 누락 월: 2016년 11월 & 12월
```

```python
# 3&4단계: 기간에 포함된 전체 월 인덱스로 생성 및 결측값 0 처리
# 01. Year = 2016, Month = 11 & 12월인 데이터프레임 만들기
left_out = {'Year':[2016, 2016], 'Month':[11, 12], 'order_id': [0, 0]}
df2 = pd.DataFrame(left_out)
df2
```

```python
# 3&4단계: 기간에 포함된 전체 월 인덱스로 생성 및 결측값 0 처리
# 02. df2 데이터프레임을 기존 데이터프레임과 합치기
orders_df = pd.concat([orders_df, df2], axis=0, ignore_index=True)
```

```python
# 3&4단계: 기간에 포함된 전체 월 인덱스로 생성 및 결측값 0 처리
# 03. 결합한 데이터프레임 연월 기준으로 정렬
orders_df = orders_df.sort_values(by=['Year', 'Month'], ascending=True, ignore_index=True)
```

```python
orders_df
```

```python
# 3&4단계: 기간에 포함된 전체 월 인덱스로 생성 및 결측값 0 처리
# 04. 전체 월 인덱스로 생성 & order_id 컬럼명 변경
orders_df.set_index(keys=['Year', 'Month'], inplace=True)
orders_df.rename(columns={'order_id':'cnt_orders'}, inplace=True)
orders_df
```

```python
# 5단계: 전월 주문건수 계산
orders_df['last_orders'] = orders_df.sort_values(by=['Year', 'Month'], ascending=True).groupby('Year')['cnt_orders'].shift(1, fill_value=0)
```

```python
orders_df
```

```python
# 6단계: 증감율 계산
# 01. 외부함수 만들기
def order_rate(a, b):
    if b == 0:
        answer = 0
    else:
        answer = (a-b)/b*100
    return answer
```

```python
# 6단계: 증감율 계산
# 02. DF 함수 적용
orders_df['growth_rate'] = orders_df.apply(lambda x: order_rate(x['cnt_orders'], x['last_orders']), axis=1)
```

```python
# 6단계: 증감율 계산
# 03. growth_Rate 값 소수점 둘째자리 반올림
orders_df['growth_rate'] = orders_df['growth_rate'].round(2)
```

```python
# 7단계: last_orders 컬럼 지우기
orders_df.drop(columns='last_orders', inplace=True)
orders_df
```
