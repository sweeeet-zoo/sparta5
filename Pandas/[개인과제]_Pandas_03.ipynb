### 최종 풀이

```python
import pandas as pd
import numpy as np
import seaborn as sns

iris= sns.load_dataset("iris")
iris

iris2 = iris.groupby('species').agg(['mean', 'std'])
iris2

flowers = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris_corr = iris[flowers].corr(method='pearson')
iris_corr = iris_corr.replace(1, 0)

iris_corr.max(axis = 0).max()
np.where(iris_corr == 0.962865431402796)
iris_corr.iloc[2:4, 2:4]

import matplotlib.pyplot as plt

sns.scatterplot(data=iris,
                x='petal_length',
                y='petal_width',
                hue='species',
                palette='husl',
                s=50)
plt.title("Iris Scatter - Petal Length and Petal Width")
plt.xlabel('Petal Length')
plt.ylabel('Petal_Width')
```

```python
# 조건 2. sepal length, sepal_width, petal_length, petal_length 4가지 변수 중 가장 상관관계가 높은 두 변수
# 01. 상관계수 구하기
flowers = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris_corr = iris[flowers].corr(method='pearson')
iris_corr
```

```python
# 조건 2. sepal length, sepal_width, petal_length, petal_length 4가지 변수 중 가장 상관관계가 높은 두 변수
# 02. 자기와의 관계 값(1)을 0으로 바꾸기
iris_corr = iris_corr.replace(1, 0)
iris_corr
```
