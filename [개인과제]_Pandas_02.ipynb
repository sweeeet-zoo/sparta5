### 최종 풀이

```python
import pandas as pd
import numpy as np

#고객 테이블
customers = pd.DataFrame({

    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],

    '이름': ['승철', '동경', '예나', '혜연', '장훈', '채운', '다연']

})

#주문 테이블
orders = pd.DataFrame({

    'cno': [1001, 1001, 1005, 1006, 1008, 1001],

    '금액': [10000, 20000, 15000, 5000, 100000, 30000]

})

orders.rename(columns = {'cno':'고객번호'}, inplace=True)
customers2 = pd.merge(customers, orders, on='고객번호', how="left")

customers2.fillna(value = {'금액':0}, inplace=True)

cond1 = customers2['금액'] == 0
customers2.loc[cond1, '구매여부'] = '미구매'

cond2 = customers2['금액'] > 0
customers2.loc[cond2, '구매여부'] = '구매'

customers2['금액'] = customers2['금액'].astype('int64')
customers2
```

### 풀이과정

```python
# 전체 데이터 확인
customers
```

```python
# 전체 데이터 확인
orders
```

```python
# 조건 1. 고객 테이블을 기준으로 주문 테이블을 합치기 (주문 금액이 없는 고객도 모두 포함)
# 01. orders 테이블의 cno 컬럼명을 '고객번호'로 바꾸기
orders.rename(columns = {'cno':'고객번호'}, inplace=True)
orders
```

```python
# 조건 1. 고객 테이블을 기준으로 주문 테이블을 합치기 (주문 금액이 없는 고객도 모두 포함)
# 02. '고객번호' 컬럼을 기준으로 두 테이블 left join으로 합치기
customers2 = pd.merge(customers, orders, on='고객번호', how="left")
customers2
```

```python
# 조건 2. 주문 금액의 결측값을 0으로 대체
customers2.fillna(value = {'금액':0}, inplace=True)
customers2
```

```python
# 조건 3. 주문금액이 있는 경우 "구매", 주문금액이 0인 경우 "미구매"로 분류하는 "구매여부" 칼럼을 생성
# loc 조건 필터를 통해 대체 값 지정하기
# 01. 주문금액이 0인 경우
cond1 = customers2['금액'] == 0
customers2.loc[cond1, '구매여부'] = '미구매'
customers2
```

```python
# 조건 3. 주문금액이 있는 경우 "구매", 주문금액이 0인 경우 "미구매"로 분류하는 "구매여부" 칼럼을 생성
# loc 조건 필터를 통해 대체 값 지정하기
# 02. 주문금액이 있는 경우
cond2 = customers2['금액'] > 0
customers2.loc[cond2, '구매여부'] = '구매'
customers2
```

```python
# 실수 타입인 '금액' 컬럼을 정수로 변환
customers2['금액'] = customers2['금액'].astype('int64')
customers2
```
