```python
df.keys()
```

```python
orders_2016 = df['orders_2016']
orders_2017 = df['orders_2017']
orders_2018 = df['orders_2018']
customers = df['customers']
payments = df['payments']
```

### orders 테이블 합치기

```python
# (1) orders 테이블이 모두 같은 컬럼으로 구성되어 있는가를 확인
print(orders_2016.columns)
print(orders_2017.columns)
print(orders_2018.columns)
```

```python
# (2) orders 테이블 수직 병합
orders = pd.concat([orders_2016, orders_2017, orders_2018], axis=0, ignore_index = True)
orders
```

```python
# 테이블 병합 및 할당 후 더블 체크!
orders.head()
```

```python
# (3) column 이름 변경
orders.rename(columns = {'order_id': '주문ID', \
                         'customer_id': '고객ID', \
                         'order_status': '주문상태', \
                         'order_purchase_timestamp': '주문시간', \
                         'order_approved_at': '승인시간', \
                         'order_delivered_carrier_date':'배송출발시간', \
                         'order_delivered_customer_date':'배송시간', \
                         'order_estimated_delivery_date':'배송예상시간'}, inplace = True)
orders
```

```python
payments.rename(columns = {'order_id':'주문ID',
                           'payment_sequential':'결제횟수',
                           'payment_type':'결제방식',
                           'payment_installments':'할부수',
                           'payment_value':'결제금액'},
                inplace = True)
payments
```

```python
customers.rename(columns =  {'customer_id':'고객ID', \
                             'customer_unique_id':'고객고유ID', \
                             'customer_zip_code_prefix':'우편번호', \
                             'customer_city':'도시', \
                             'customer_state':'주'}, inplace = True)
customers
```

### 데이터 병합

```python
customers2 = pd.merge(orders, customers, how = 'inner', on = '고객ID')
customers2
```

```python
payments2 = pd.merge(payments, orders, how ='inner', on = '주문ID')
payments2
```

```python
customers2.to_csv("customers2.csv", index = True, encoding = "UTF-8-SIG")
```

```python
payments2.to_excel("payments2.xlsx", index = True)
```

```python
payments2.to_csv("payments2.csv", index = True, encoding = "UTF-8-SIG")
```
