### 2-4) 결측값 처리(삽입)

```python
orders
```

```python
# 배송시간 결측값 처리
orders["배송출발시간"] = pd.to_datetime(orders["배송출발시간"])
orders["배송시간"] = pd.to_datetime(orders["배송시간"])
```

```python
orders.dtypes
```

```python
mean_delivery = (orders["배송시간"] - orders["배송출발시간"]).mean()
mean_delivery
```

```python
orders["배송추정시간"] = orders["배송출발시간"] + mean_delivery
orders
```

```python
orders[orders["배송시간"].isna()]
```

```python
# orders copy 해두기
orders2 = orders.copy()
```

```python
orders["배송시간"].fillna(orders["배송추정시간"], inplace = True)
```

```python
orders2.isna().sum()
```

```python
orders2[orders["배송시간"].notna()]
```

```python
orders2.loc[orders["배송시간"].notna(), "배송추정시간"] = orders["배송시간"]
```

```python
orders2.rename(columns = {"배송추정시간":"배송(추정)시간"}, inplace = True)
orders2.isna().sum()
```
