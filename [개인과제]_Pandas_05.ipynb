```python
# 조건 1. 전체 데이터 결측값 처리
df.dropna(axis=0, how='any', inplace=True)
df
```

```python
# 조건 2. Quantity, UnitPrice 0 이하 제거
cond1 = (df.Quantity > 0) & (df.UnitPrice > 0)
df = df.loc[cond1]
df
```

```python
# 조건 3. 상품코드가 일반적이지 않은 경우
# 문자로만 구성된 경우 = 숫자를 한개 이상 포함하지 않는 경우
# 01. 임의로 숫자를 지정하여 해당 문자열에 몇개가 들어있는지 확인하기
df['StockCode'].str.count('2')
```

```python
# 조건 3. 상품코드가 일반적이지 않은 경우
# 문자로만 구성된 경우 = 숫자를 한개 이상 포함하지 않는 경우
# 02. 상품코드의 문자열 원소를 i로 정의하고 1번 과정을 반복하는 함수 만들기

# 1) 수식 테스트
answer = 0
for i in range(0, 10):
    num = df['StockCode'].str.count(str(i))
    answer += num
print(answer)
```

```python
# 조건 3. 상품코드가 일반적이지 않은 경우
# 문자로만 구성된 경우 = 숫자를 한개 이상 포함하지 않는 경우
# 02. 상품코드의 문자열 원소를 i로 정의하고 1번 과정을 반복하는 함수 만들기

# 2) 함수 테스트
def cnt_code(x):
    answer = 0
    for i in range(0, 10):
        num = x.str.count(str(i))
        answer += num
    return answer

cnt_code(df['StockCode'])
```

```python
# 조건 3. 상품코드가 일반적이지 않은 경우
# 문자로만 구성된 경우 = 숫자를 한개 이상 포함하지 않는 경우
# 03. 데이터 프레임에 함수 적용하기
df["cnt_num"] = cnt_code(df['StockCode'])
```

```python
# 조건 3. 상품코드가 일반적이지 않은 경우
# 문자로만 구성된 경우 = 숫자를 한개 이상 포함하지 않는 경우
# 04. 숫자를 한개 이상 포함하는 경우 검색
cond2 = df['cnt_num']>=1
df.loc[cond2]
```

```python
# 조건 3. 상품코드가 일반적이지 않은 경우
# 문자로만 구성된 경우 = 숫자를 한개 이상 포함하지 않는 경우
# 05. 검색 결과를 데이터프레임에 덮어쓰기
cond2 = df['cnt_num']>=1
df = df.loc[cond2]
df
```

```python
# 조건 4. Quantity, UnitPrice 컬럼 IQR 기반 방법으로 이상치 제거
# IQR (Inter Quantitle Range) = Q3 (3사분위 수) - Q1 (1사분위수)
# Quantity컬럼과 UnitPrice컬럼에서, lower_bound(Q1 - 1.5 * IQR) 보다 작거나 upper_bound( Q3 + 1.5 * IQR) 보다 큰 경우를 이상치로 정의
# 01. IQR & 1/3사분위 값 정의

# Quantity 컬럼
q_Q1 = df['Quantity'].quantile(.25)
q_Q3 = df['Quantity'].quantile(.75)
q_IQR = q_Q3 - q_Q1

# UnitPrice 컬럼
u_Q1 = df['UnitPrice'].quantile(.25)
u_Q3 = df['UnitPrice'].quantile(.75)
u_IQR = q_Q3 - q_Q1
```

```python
# 조건 4. Quantity, UnitPrice 컬럼 IQR 기반 방법으로 이상치 제거
# IQR (Inter Quantitle Range) = Q3 (3사분위 수) - Q1 (1사분위수)
# Quantity컬럼과 UnitPrice컬럼에서, lower_bound(Q1 - 1.5 * IQR) 보다 작거나 upper_bound( Q3 + 1.5 * IQR) 보다 큰 경우를 이상치로 정의
# 02. 각 컬럼의 lower_bound와 upper_bound 정의

# Quantity 컬럼
q_lower_bound = q_Q1 - 1.5 * q_IQR
q_upper_bound = q_Q3 + 1.5 * q_IQR

# UnitPrice 컬럼
u_lower_bound = u_Q1 - 1.5 * u_IQR
u_upper_bound = u_Q3 + 1.5 * u_IQR
```

```python
u_upper_bound
```

```python
# 조건 4. Quantity, UnitPrice 컬럼 IQR 기반 방법으로 이상치 제거
# IQR (Inter Quantitle Range) = Q3 (3사분위 수) - Q1 (1사분위수)
# Quantity컬럼과 UnitPrice컬럼에서, lower_bound(Q1 - 1.5 * IQR) 보다 작거나 upper_bound( Q3 + 1.5 * IQR) 보다 큰 경우를 이상치로 정의
# 03. value < lower_bound or value > upper_bound 조건 정의
cond3 = (df['Quantity'] < q_lower_bound) | (df['Quantity'] > q_upper_bound)
cond4 = (df['UnitPrice'] < u_lower_bound) | (df['UnitPrice'] > u_upper_bound)
```

```python
# 조건 4. Quantity, UnitPrice 컬럼 IQR 기반 방법으로 이상치 제거
# IQR (Inter Quantitle Range) = Q3 (3사분위 수) - Q1 (1사분위수)
# Quantity컬럼과 UnitPrice컬럼에서, lower_bound(Q1 - 1.5 * IQR) 보다 작거나 upper_bound( Q3 + 1.5 * IQR) 보다 큰 경우를 이상치로 정의
# 04. 해당 조건을 충족하는  조회
df.loc[cond3 | cond4]
```

```python
# 조건 4. Quantity, UnitPrice 컬럼 IQR 기반 방법으로 이상치 제거
# IQR (Inter Quantitle Range) = Q3 (3사분위 수) - Q1 (1사분위수)
# Quantity컬럼과 UnitPrice컬럼에서, lower_bound(Q1 - 1.5 * IQR) 보다 작거나 upper_bound( Q3 + 1.5 * IQR) 보다 큰 경우를 이상치로 정의
# 05. 조건 충족 행 제거

# 1) 해당 행의 인덱스 출력
df.loc[cond3 | cond4].index
```

```python
# 조건 4. Quantity, UnitPrice 컬럼 IQR 기반 방법으로 이상치 제거
# IQR (Inter Quantitle Range) = Q3 (3사분위 수) - Q1 (1사분위수)
# Quantity컬럼과 UnitPrice컬럼에서, lower_bound(Q1 - 1.5 * IQR) 보다 작거나 upper_bound( Q3 + 1.5 * IQR) 보다 큰 경우를 이상치로 정의
# 05. 조건 충족 행 제거

# 2) 조건 충족하는 행의 인덱스 저장
iqr_index = df.loc[cond3 | cond4].index
```

```python
# 조건 4. Quantity, UnitPrice 컬럼 IQR 기반 방법으로 이상치 제거
# IQR (Inter Quantitle Range) = Q3 (3사분위 수) - Q1 (1사분위수)
# Quantity컬럼과 UnitPrice컬럼에서, lower_bound(Q1 - 1.5 * IQR) 보다 작거나 upper_bound( Q3 + 1.5 * IQR) 보다 큰 경우를 이상치로 정의
# 05. 조건 충족 행 제거

# 3) iqr_index 행 제거 및 덮어쓰기
df.drop(iqr_index, inplace=True)
df
```
