### 최종 풀이

```python
!pip install -U finance-datareader
import FinanceDataReader as fdr

import pandas as pd

samsung_fdr = fdr.DataReader('005930', '2019-01-01', '2024-12-16')
samsung_fdr

samsung_fdr = samsung_fdr.reset_index(col_fill='Date_fdr')
samsung_fdr['Year'] = samsung_fdr['Date'].dt.year
samsung_fdr['Month'] = samsung_fdr['Date'].dt.month
samsung_fdr

# 4-1) 조건 충족
samsung_mean = samsung_fdr.groupby(['Year', 'Month'])['Close'].mean().reset_index()
samsung_mean

# 4-2 & 3) 조건 충족
samsung_mean["Last_Close"] = samsung_mean.sort_values(
                              by='Month', ascending=True
                              ).groupby('Year')['Close'].shift(1, fill_value=0)

def finance_rate(a, b): # 조건에 사용할 외부함수 생성
    if b == 0:
        answer = 0
    else:
        answer = (a-b)/b*100
    return answer

samsung_mean['전월대비(%)'] = samsung_mean.apply(lambda x: finance_rate(x['Close'], x['Last_Close']), axis=1).round(1)
samsung_mean['전년동월대비(%)'] = samsung_mean.apply(lambda x: finance_rate(x['Close'], x['Last_Year_Close']), axis=1).round(1)

# 최종 데이터프레임 출력
samsung_mean
```

```python
# 조건 2. 월평균 종가 구하기
# 01. 날짜별 연&월 추출하기
import datetime as dt

samsung_fdr['Year'] = samsung_fdr['Date'].dt.year
samsung_fdr['Month'] = samsung_fdr['Date'].dt.month
samsung_fdr
```

```python
# 조건 2. 월평균 종가 구하기
# 02. 연도&월별 그룹화 및 평균 종가 구하기
samsung_mean = samsung_fdr.groupby(['Year', 'Month'])['Close'].mean().reset_index()
samsung_mean
```

#### 전월대비 증감률

```python
# 조건 3. 평균 종가의 전월대비 증감률 구하기
# 01. Month 열을 기준으로 행을 하나씩 내리기
# - 1) 임의의 그룹 기준 만들어 주기: Company = Samsung
# samsung_mean.insert(2, 'Company', 'Samsung')
```

```python
# 조건 3. 평균 종가의 전월대비 증감률 구하기
# 01. Month 열을 기준으로 행을 하나씩 내리기
# - 2) Year & Month groupby & shift 활용(결측치는 0으로 대체)
samsung_mean["Last_Close"] = samsung_mean.sort_values(
                              by=['Year','Month'], ascending=True
                              ).groupby('Year')['Close'].shift(1, fill_value=0)
samsung_mean
```

```python
# 조건 3. 평균 종가의 전월대비 증감률 구하기
# 02. 전월대비 증감률을 구하는 외부함수 만들기
def finance_rate(a, b):
    if b == 0:
        answer = 0
    else:
        answer = (a-b)/b*100
    return answer
```

```python
# 조건 3. 평균 종가의 전월대비 증감률 구하기
# 03. 전월대비(%) 컬럼을 생성하고 외부함수 적용하기
samsung_mean['전월대비(%)'] = samsung_mean.apply(lambda x: finance_rate(x['Close'], x['Last_Close']), axis=1)
samsung_mean
```

```python
# 조건 3. 평균 종가의 전월대비 증감률 구하기
# 04. 전월대비(%) 값의 소수점 첫째자리까지 반올림하기
samsung_mean['전월대비(%)']=samsung_mean['전월대비(%)'].round(1)
samsung_mean
```

#### 전년동월대비 증감률

```python
# 조건 4. 평균 종가의 전년동월대비 증감률 구하기
# 01. Year 열을 기준으로 행을 하나씩 내리기
# -> Year & Month groupby & shift 활용(결측치는 0으로 대체)
samsung_mean["Last_Year_Close"] = samsung_mean.sort_values(
                              by=['Year','Month'], ascending=True
                              ).groupby('Month')['Close'].shift(1, fill_value=0)
samsung_mean
```

```python
# 조건 4. 평균 종가의 전년동월대비 증감률 구하기
# 02. 전년동월대비(%) 컬럼을 생성하고 외부함수 적용하기
samsung_mean['전년동월대비(%)'] = samsung_mean.apply(lambda x: finance_rate(x['Close'], x['Last_Year_Close']), axis=1)
samsung_mean
```

```python
# 조건 4. 평균 종가의 전년동월대비 증감률 구하기
# 03. 전년동월대비(%) 값의 소수점 첫째자리까지 반올림하기
samsung_mean['전년동월대비(%)']=samsung_mean['전년동월대비(%)'].round(1)
samsung_mean
```
