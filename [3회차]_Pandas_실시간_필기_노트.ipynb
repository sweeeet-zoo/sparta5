### 1.1 apply(func, axis=0)
    axis = 0 -> 행 기준
    axis = 1 -> 열 기준

```python
# 데이터프레임 형성
df = pd.DataFrame({
    'value': [10, 20, 30],
    'operation': ['square', 'double', 'negate']
})
df
```

```python
# operation 값에 따라 계산
def perform_operation(row):
    if row['operation'] == 'square':
        return row['value'] ** 2
    elif row['operation'] == 'double':
        return row['value'] * 2
    elif row['operation'] == 'negate':
        return -row['value']
```

```python
df['result'] = df.apply(perform_operation, axis=1)
df
```

2. datetime 형태로 type 변환

```python
payments2["주문시간"] = pd.to_datetime(payments2["주문시간"])
payments2["주문시간"]
```

```python
payments2["주문연도"] = payments2["주문시간"].dt.year
payments2["주문연도"]
```

```python
payments2['주문월'] = payments2['주문시간'].dt.month
```

```python
payments2
```

3. "주문연월" 컬럼 생성하기

```python
# 첫번째 방법 [문자열 결합]
payments2["주문연도"].astype(str) + "-" + payments2["주문월"].astype(str)
```

```python
# 두번째 방법 [dt.strftime 이용]
# 함수만 실행할 경우 output이 나오지만, 컬럼명을 지정해줄 경우 '할당' 기능이 실행되어 output은 출력되지 않는다.
# payments2["주문시간"].dt.strftime("%Y-%m") -> output 출력 O
payments2["주문연월"] = payments2["주문시간"].dt.strftime("%Y-%m") # -> ouput 출력 X
```

3) 그룹화

```python
# payments2에서 "결제방식"별 "결제금액" 평균을 구해보자 (pay_mean) 생성
pay_mean = payments2.groupby('결제방식')['결제금액'].mean()
```

```python
# payments2에서 "결제방식"별 "결제금액" 평균, 합계 구해보자 (pay_groupby) 생성
pay_groupby = payments2.groupby('결제방식')['결제금액'].agg(["mean","sum"])
```

```python
# 인덱스를 컬럼으로 바꾸기
pay_mean = pay_mean.reset_index()
pay_mean
```

```python
pay_groupby.reset_index(inplace = True)
pay_groupby
```

4) 피봇팅

```python
# payments2에서 주문연월별 주문상태 및 결제방식에 따른 평균 결제 금액, 고유 count을 알아보자. 총합계 출력 (result)
result = payments2.pivot_table(values = "결제금액",
                      index = ["주문상태", "결제방식"],
                      columns = "주문연월",
                      aggfunc = ["mean", "count"],
                      fill_value = 0,
                      margins = True)
```

```python
result.reset_index(inplace=True)
```

```python
result.to_csv("result.csv")
```
